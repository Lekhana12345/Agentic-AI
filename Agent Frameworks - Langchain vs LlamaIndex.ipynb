# LangChain with Google Gemini

## Objective
To integrate Google Gemini with LangChain and perform a simple query.

## Description
This notebook demonstrates:
- Using Gemini inside LangChain
- Creating an LLM wrapper
- Running a query using LangChain interface

```python
import google.generativeai as genai
```

- Imports Gemini SDK for API interaction.

```python
genai.configure(api_key="YOUR_API_KEY")
```

- Configures API access for Gemini services.

```python
from langchain_google_genai import ChatGoogleGenerativeAI
```

- Imports LangChain wrapper for Gemini.
- Allows Gemini to be used like a standard LangChain LLM.

```python
llm = ChatGoogleGenerativeAI(
    model="gemini-3.1-pro-preview",
    google_api_key="YOUR_API_KEY"
)
```

### Explanation

- Initializes the Gemini model inside LangChain.
- `ChatGoogleGenerativeAI` acts as a bridge between LangChain and Gemini.
- Enables use of chains, agents, and tools.

Parameters:
- model → Specifies which Gemini model to use
- google_api_key → Authentication key

```python
response = llm.invoke("Explain AI agents in simple terms")
```

### Explanation

- Sends a query to the Gemini model through LangChain.
- `.invoke()` is the standard method in LangChain to call the model.
- Returns a structured response object.

```python
print(response.content)
```

- Extracts and prints the actual text output from the response object.
