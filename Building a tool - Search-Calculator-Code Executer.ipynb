# ReAct Agent using LangChain and Gemini

## Objective
To build a tool-augmented AI agent using the ReAct (Reasoning + Acting) framework with LangChain and Google Gemini.

## Description

This notebook demonstrates:
- LLM-based reasoning (Gemini)
- Tool integration (Search, Calculator, Python)
- Autonomous decision-making using ReAct

## Key Concept: ReAct Framework

ReAct combines:
- Reasoning (LLM thinking step-by-step)
- Acting (using tools)

Workflow:
Thought → Action → Observation → Final Answer

This enables real-world task solving.

```python
from langchain_google_genai import ChatGoogleGenerativeAI
from langchain.agents import initialize_agent, Tool
from langchain.utilities import SerpAPIWrapper
from langchain_experimental.tools import PythonREPLTool
import os
```

### Explanation

- `ChatGoogleGenerativeAI` → Gemini LLM integration
- `initialize_agent` → Creates AI agent
- `Tool` → Defines external tools
- `SerpAPIWrapper` → Web search tool
- `PythonREPLTool` → Executes Python code

These components enable **tool-augmented reasoning**.

```python
# 🔑 API Keys
os.environ["SERPAPI_API_KEY"] = "YOUR_SERP_API_KEY"
```

### Explanation

- Sets API key for SerpAPI (used for web search)
- Required for real-time information retrieval

⚠️ Best Practice:
Use environment variables instead of hardcoding keys.

```python
# 🤖 Gemini LLM
llm = ChatGoogleGenerativeAI(
    model="gemini-3.1-pro-preview",
    google_api_key="YOUR_GEMINI_API_KEY"
)
```

### Explanation

- Initializes Gemini LLM inside LangChain
- Acts as the **reasoning engine**

This model:
- Decides which tool to use
- Generates reasoning steps
- Produces final answers

```python
# 🌐 Search Tool
search = SerpAPIWrapper()
```

### Explanation

- Enables real-time web search
- Used when the agent needs external information

Example:
"latest news" → triggers search tool

```python
# 🧮 Calculator Tool
def calculator_tool(query):
    try:
        return str(eval(query))
    except:
        return "Error"
```

### Explanation

- Evaluates mathematical expressions
- Uses Python `eval()` for computation

⚠️ Note:
- Not safe for production (can execute arbitrary code)
- Should be replaced with secure parser

```python
# 💻 Python Tool
python_tool = PythonREPLTool()
```

### Explanation

- Executes Python code dynamically
- Useful for:
  - data analysis
  - scripting
  - calculations

Acts as a **general-purpose execution tool**

```python
# 🧰 Tools Setup
tools = [
    Tool(name="Search", func=search.run, description="Search the web"),
    Tool(name="Calculator", func=calculator_tool, description="Math calculations"),
    Tool(name="PythonREPL", func=python_tool.run, description="Run Python code"),
]
```

### Explanation

Defines the tools available to the agent:

- Search → for real-world data
- Calculator → for math problems
- PythonREPL → for code execution

The agent decides:
- which tool to use
- when to use it

```python
# 🧠 Initialize Agent
agent = initialize_agent(
    tools,
    llm,
    agent="zero-shot-react-description",
    verbose=True
)
```

### Explanation

- Creates a **ReAct agent**
- "zero-shot-react-description" means:
  - No training needed
  - Uses tool descriptions to decide actions

`verbose=True`:
- Shows reasoning steps:
  Thought → Action → Observation

```python
# 🚀 Run Agent
while True:
    query = input("Ask: ")
    response = agent.run(query)
    print("\nFinal Answer:", response)
```

### Explanation

Execution loop:

1. User inputs query
2. Agent processes using ReAct:
   - Think (LLM reasoning)
   - Act (tool usage)
   - Observe (tool output)
3. Final answer is generated

This creates an **interactive AI system**

## System Architecture

### Components

1. Input Layer
   - User query

2. LLM Layer (Gemini)
   - Reasoning and decision making

3. Tool Layer
   - Search (web)
   - Calculator (math)
   - Python REPL (execution)

4. Agent Layer (ReAct)
   - Controls reasoning loop

### Workflow

Query → LLM → Tool Selection → Tool Execution → Observation → Final Answer
